```python
def wordlist(name):
    list2 = pd.read_csv('./transaction-data/'+name+'.csv')
    list2.columns=['auth_id','Date','Vendor', 'Amount','Location']
    z2=list2['Vendor'].str.cat(sep=',')
    wordlist = re.split(',|/|-| ', z2)
    wordfreq = []
    for w in wordlist:
        wordfreq.append(wordlist.count(w))
    print("Pairs\n" + str(zip(wordlist, wordfreq)))
```

```python
wordlist('user-2')
```
