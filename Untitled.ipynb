```python
from collections import Counter
def MostPop(i):
    list2 = pd.read_csv('./transaction-data/user-'+str(i)+'.csv')
    list2.columns=['auth_id','Date','Vendor', 'Amount','Location']
    z2=list2['Vendor'].str.cat(sep=',')
    wordlist = re.split(',|/|-| ', z2)
    
    wordfreq = {}
    for word in wordlist:
        if word not in wordfreq:
            wordfreq[word] = 0 
        wordfreq[word] += 1
    if '' in wordfreq: del wordfreq['']
    return Counter(wordfreq).most_common(20)  
```

```python

def compare(i,j):
    li1= []
    li2 = []
    for letter, count in MostPop(i):
        li1.append(letter)
    for letter, count in MostPop(j):
        li2.append(letter)
    common_elements = list(set(li1).intersection(set(li2)))
    return len(common_elements)
```

```python
print compare(1,2)
```

```python
print compare(1,3)
```
